Para evitar la forma hiperbólica de las curvas, podemos elevar al cuadrado cada una de las variables, y así obtenemos 3 rectas que nos permiten usar el método Dix. A continuación vemos la gráfica resultante:

```python
##Interceptos y pendientes elevados al cuadrado.
import matplotlib.pyplot as plt
import numpy as np

# Datos
distancia = list(range(3, 61, 3))
distancia_cuadrado = [x ** 2 for x in distancia]  # Elevando al cuadrado los valores de distancia

reflexion1 = [x ** 2 for x in [30.57, 30.76, 31.08, 31.53, 32.10, 32.78, 33.56, 34.44,
                               35.42, 36.47, 37.61, 38.81, 40.08, 41.40, 42.78, 44.21,
                               45.68, 47.18, 48.73, 50.30]]

reflexion2 = [x ** 2 for x in [42.32, 42.45, 42.64, 42.90, 43.23, 43.64, 44.11, 44.65,
                               45.25, 45.91, 46.64, 47.42, 48.25, 49.14, 50.07, 51.05,
                               52.07, 53.13, 54.23, 55.36]]

reflexion3 = [x ** 2 for x in [50.12, 50.19, 50.31, 50.47, 50.68, 50.93, 51.22, 51.56,
                               51.94, 52.36, 52.83, 53.33, 53.87, 54.45, 55.06, 55.71,
                               56.39, 57.11, 57.85, 58.63]]

# Calcular la pendiente e intercepto de la regresión lineal para cada reflexión
pendiente_reflexion1, intercept_reflexion1 = np.polyfit(distancia_cuadrado, reflexion1, 1)
pendiente_reflexion2, intercept_reflexion2 = np.polyfit(distancia_cuadrado, reflexion2, 1)
pendiente_reflexion3, intercept_reflexion3 = np.polyfit(distancia_cuadrado, reflexion3, 1)

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.plot(distancia_cuadrado, reflexion1, label=f'Pendiente - Reflexión 1: {pendiente_reflexion1:.2f}')
plt.plot(distancia_cuadrado, reflexion2, label=f'Pendiente - Reflexión 2: {pendiente_reflexion2:.2f}')
plt.plot(distancia_cuadrado, reflexion3, label=f'Pendiente - Reflexión 3: {pendiente_reflexion3:.2f}')

# Línea de intercepto
plt.plot([0], [intercept_reflexion1], marker='o', markersize=5, label=f'Intercepto ^2 - Reflexión 1: {intercept_reflexion1:.2f}')
plt.plot([0], [intercept_reflexion2], marker='o', markersize=5, label=f'Intercepto ^2 - Reflexión 2: {intercept_reflexion2:.2f}')
plt.plot([0], [intercept_reflexion3], marker='o', markersize=5, label=f'Intercepto ^2 - Reflexión 3: {intercept_reflexion3:.2f}')

# Etiquetas de los ejes y leyenda
plt.xlabel('Distancia (m^2)')
plt.ylabel('Velocidad de reflexión (ms^2)')
plt.legend()

# Establecer límites personalizados para los ejes X e Y
plt.xlim(0, 4000)  # Límite X hasta 3600
plt.ylim(500, 3500)  # Límite Y hasta 3500

# Título del gráfico
plt.title('tiempo vs distancia ')

# Mostrar el gráfico
#plt.grid()
plt.grid(True, which='both', alpha=0.2)
plt.show()
```

Para cada recta (reflexión 1 , reflexión 2 , reflexión 3) se encontraron los interceptos con el eje y, los cuales corresponden a **to**. Esto nos ayudará más adelante a calcular el **espesor** para cada estrato.

- **Intercepto al cuadrado de la reflexión 1** = 930.28

- **Intercepto al cuadrado de la reflexión 2** = 1789.15

- **Intercepto al cuadrado de la reflexión 3** = 2510.10

También se encontraron las pendientes correspondientes a cada recta:

- **Pendiente de la reflexión 1** = 0.44

- **Pendiente de la reflexión 2** = 0.35

- **Pendiente de la reflexión 3** = 0.26

Utilizando las pendientes de las rectas obtenidas en la gráfica anterior, encontramos las velocidades medias cuadráticas:

> **m = 1/(vrms)^2**

> **vrms = 1/(m)^1/2**

- **Vrsm_1**: 1.51 milisegundos

- **Vrsm_2**: 1.69 milisegundos

- **Vrsm_3**: 1.96 milisegundos

Debemos recordar que para aplicar este método asumimos que la distancia entre emisor y receptores es pequeña en comparación con la profundidad de los reflectores **(h>>x)**, y que son trayecotrias aproximadamene verticales.


Luego de encontrar las velocidades, aplicamos el método de dix, que consiste en estimar las velocidades Vn hasta Vn+1 (segundos) y los espesores h (metros) correspondientes a partir de los tiempos:

 **V(n)^2 =[(Vrsm^2).to(n)-Vrsm(n-1)^2]/to(n)-to(n-1)**



Sabemos que la velocidad de la primera reflexión es

- **Velocidad de la capa 1** = 1.50 milisegundos


```python
##Tenemos vrsm por cada capa
Vrsm_1 = 1.50
Vrsm_2 = 1.69
Vrsm_3 = 1.96
##Tenemos intecepto por cada capa elevado al cuadrado
intercepto_1 = 930.28**(1/2)
intercepto_2 = 1789.15**(1/2)
intercepto_3= 2510.10**(1/2)
v1 = 1.50
v2 = (((Vrsm_2*intercepto_2)-(Vrsm_1*intercepto_1))/(intercepto_2-intercepto_1))**(1/2)
v3 = (((Vrsm_3*intercepto_3)-(Vrsm_2*intercepto_2))/(intercepto_3-intercepto_2))**(1/2)
print("La velocidad de la capa 1 es", v1, "(m/ms)")
print("La velocidad de la capa 2 es", round(v2, 2), "(m/ms)")
print("La velocidad de la capa 3 es", round(v3, 2), "(m/ms)")
```

```python
print("La velocidad de la capa 1 es", round(v1*1000,2), "(m/s)")
print("La velocidad de la capa 2 es", round(v2*1000,2), "(m/s)")
print("La velocidad de la capa 3 es", round(v3*1000,2), "(m/s)")
```

Ahora para encontrar los espesores de capa aplicaremos la siguiente fórmula:
- **hn=vn((to(n)-to(n-1)/2)**

Ya conocemos las velocidades de cada capa y los interceptos, podemos reemplazar:

```python
h1 = ((v1*(intercepto_1)/2))
h2= ((v2*(intercepto_2-intercepto_1)/2))
h3 = ((v3*(intercepto_3-intercepto_2)/2))

print("El espesor 1 es =", round(h1,2), "metros")
print("El espesor 2 es =", round(h2,2), "metros")
print("El espesor 3 es =", round(h3,2), "metros")
```

**b)** Para cada posición (distancia x) estime la velocidad media utilizando el método del normal-move-out y compare los resultados obtenidos en el literal anterior. Asuma capas horizontales.

El normal move out es un proceso importante en el procesamiento de datos sismicos de reflexión. Permite al final obtener un pulso mayor que representa de mejor manera un reflector en particular.

El tnmo es la diferencia entre el tiempo de reflexión en el sitio de disparo y el tiempo que le toma a la onda reflejada para llegar al receptor que se encuentra a distancia
.
**Para la curva de reflexión 1**
Tomamos como punto de tiempo inicial  
- **t0=**30.500 ms


```python
t01 = 30.500  # milisegundos
reflexion1 = [30.57, 30.76, 31.08, 31.53, 32.10, 32.78, 33.56, 34.44, 35.42, 36.47,
              37.61, 38.81, 40.08, 41.40, 41.40, 42.78, 44.21, 45.68, 47.18, 48.73, 50.30]

# Inicializa una lista para almacenar los resultados de las restas
resultados = []

# Recorre la lista de reflexion1
for valor in reflexion1:
    resultado = valor - t01
    resultados.append(resultado)

# Imprime los resultados
for i, resultado in enumerate(resultados):
    print(f"Resultado {i + 1}: {resultado} ms")
```

```python
t01 = 30.500  # milisegundos
reflexion1 = [30.57, 30.76, 31.08, 31.53, 32.10, 32.78, 33.56, 34.44, 35.42, 36.47,
              37.61, 38.81, 40.08, 41.40, 41.40, 42.78, 44.21, 45.68, 47.18, 48.73, 50.30]
x = list(range(3, 61, 3))

# Inicializa una lista para almacenar los resultados de las restas
resultados = []

# Recorre la lista de reflexion1 y calcula los resultados
for valor in reflexion1:
    resultado = valor - t01
    resultados.append(resultado)

# Inicializa una lista para almacenar los valores de v1
v1_values = []

# Calcula v1 para cada valor de resultado y x
for r, distancia in zip(resultados, x):
    v1 = (distancia) / (2 * r * t01) ** (1/2)
    v1_values.append(v1)

# Imprime los resultados de v1
for i, v1 in enumerate(v1_values):
    print(f"distancia={x[i]} normal move out={resultados[i]}: velocidad para por medio del normal move out en la capa 1 {v1}")
```

Esta fórmula calcula el angulo de buzamiento por cada distancia:

```python
import math

# Datos
tdmo = [2.770, 2.607, 2.470, 2.290, 2.100, 1.890, 1.660, 1.410, 1.160, 0.880, 0.590, 0.300]
distancia = [36, 33, 30, 27, 24, 21, 18, 15, 12, 9, 6, 3]
v_p2 =  1.9114835325634763

# Lista para almacenar los ángulos
beta_p2_b = []

# Calcular el ángulo en cada distancia
for i in range(len(tdmo)):
    beta = math.asin((v_p2 * tdmo[i]) / (2 * distancia[i]))
    beta_p2_b.append(math.degrees(beta))  # Convertir el ángulo a grados

# Imprimir los ángulos
for i, beta in enumerate(beta_p2_b):
    print(f'β a {distancia[i]} m: {beta:.2f}°')

```

Con los ángulos obtenidos por medio del método **Tdmo** y comparando el ángulo β por medio del método **tmin/t0**, podemos encontrar una variación en los resultados entre ambos métodos:
- Por medio del método tmin/t0, el ángulo fue de 6.38°. Para calcular este valor solo se tienen en cuenta tmin y t0, y no la forma del reflector a, diferencia del tdmo, en donde sí se tiene en cuenta para cada distancia. Esto nos hace considerar que la geometría del reflector en capas con buzamientos bajos no es tan significativa a diferencia en donde hay ángulos de inclinación más grandes, por lo tanto el cálculo de β por el método de tmin/t podría ser una buena aproximación al angulo de inclnación real de la frontera.

Posteriomente se cálcula **j** para cada distancia teniendo en cuenta la siguiente fórmula:

- t^2 = 4j^2+ x^2- 4jx*sin(β) / v

Ahora despejando **j** obtenemos:

- t^2*v-4j^2 = x^2- 4jx*sin(β)

- 4j^2- 4jx*sin(β)+(t^2v+x^2) =0

En donde se dan como soluciones dos j (una positiva y otra negativa), tomando la positiva y quedando la fórmula así:

- **j** = 1/2* (x*sin(β)+((x^2*sin(β)-x^2+t^2*v))**1/2

Dos geófonos simétricos en relación con el sitio de disparo (SP) registraron los siguientes tiempos de llegada al mismo nivel reflector: t(G-x) = 0.8 s y t(Gx) = 0.7 s. La distancia del SP a cada geófono es de 300 m y la velocidad del medio es 1200 m/s. Calcular el espesor j de la capa y el buzamiento aparente β
de la superficie reflectora a lo largo del perfil de reflexión (j es la distancia perpendicular entre el sitio de disparo y la superficie reflectora)

Es importante decir que para este ejercicio el estrato eta inclinado, po eso aplicamos tdmo.

```python
#datos
tG_x =	0.8 #segundos
tGx  = 	0.7 #segundos
V =	1200 #metros/#segundos
x	= 300 #metros
```

Sabemos que
- **|T(-X)-T (+X)|**

- **β** = arcseno(v*tdmo/2x)
- **Tx**= (4*j^2+x^2-4*j*x*sen(b))*1/2

Sabemos que para calcular **j** se tiene en cuenta la fórmula anteriormente mencionada

Ahora despejando **j** obtenemos:

- t^2*v-4j^2 = x^2- 4jx*sin(β)

- 4j^2- 4jx*sin(β)+(t^2v+x^2) =0

En donde se dan como soluciones dos j (una positiva y otra negativa), tomando la positiva y quedando la fórmula así:

- **j** = 1/2* (x*sin(β)+((x^2*sin(β)-x^2+t^2*v))**1/2

```python
import math

# Definir las variables tG_x, tGx, v_p2 y x
tG_x = 0.8
tGx = 0.7
v_p4 = 1200
x = 300

# Calcular TDMO_P4
TDMO_P4 = abs(tG_x - tGx)

# Imprimir el resultado de TDMO_P4
print("El valor de TDMO_P4 es:", TDMO_P4)

# Calcular beta
beta_p4 = math.degrees(math.asin((v_p4 * TDMO_P4) / (2 * x)))
print("El ángulo es:", beta_p4, "°")

# Calcular j
j_4 = 0.5 * (x * math.sin(math.radians(beta_p4)) + math.sqrt((x ** 2) * (math.sin(math.radians(beta_p4)) ** 2) - x ** 2 + (tGx ** 2) * (v_p4 ** 2)))

# Imprimir el resultado de j
print("El valor de j es:", j_4, "metros")
```
